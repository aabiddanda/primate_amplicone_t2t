```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from glob import glob
from pathlib import Path
from tqdm import tqdm
%matplotlib inline 
```

```python

dfs = []
species = ['mGorGor1_gorilla', 'mPanPan1_bonobo', 'mPanTro3_chimp', 'mPonAbel_orangutan_sumatra', 'mPonPyg2_orangutan_borneo']
for s in tqdm(species):
    for f in glob(f'../results/{s}/amplicone_results/*.Ampliconic_Summary.txt'):
        fp = Path(f)
        name = fp.name.split('.')[0]
        df = pd.read_csv(f, sep="\t")
        df['species'] = s
        df['name'] = name
        dfs.append(df)

meta_amplicone_df = pd.concat(dfs)
meta_amplicone_df
```

```python
# Really 
meta_amplicone_df.groupby(['species', 'GeneFamily'])['CopyNumber(MAP=1)'].agg(lambda x: np.nanmean(x)).reset_index().sort_values('GeneFamily')
```
